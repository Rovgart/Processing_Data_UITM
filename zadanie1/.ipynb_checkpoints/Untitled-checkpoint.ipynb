```python
import numpy as np
```

```python
oceny = np.array([3.0, 4.0, 5.0, 3.5, 4.5])
print(oceny)
```

```python
type(oceny)
```

```python
oceny.shape
```

```python
print(oceny[0])
print(oceny[2])
```

```python
print(oceny[1:4])
```

```python
np.mean(oceny)
```

```python
mediana = np.median(oceny)
minimum = np.min(oceny)
maksimum = np.max(oceny)
suma = np.sum(oceny)
odchylenie = np.std(oceny)

print(f"Mediana: {mediana}")
print(f"Minimum: {minimum}")
print(f"Maksimum: {maksimum}")
print(f"Suma: {suma}")
print(f"Odchylenie standardowe: {odchylenie:.2f}")
```
