```python
wyniki = np.array([45, 88, 52, 30, 95, 67, 48, 72, 55, 12])
```

```python
poprawione_wyniki = wyniki + 5
```

```python
poprawione_wyniki = np.clip(poprawione_wyniki, 0, 100)
```

```python
nowa_srednia = np.mean(poprawione_wyniki)
```

```python
print(f"Poprawione wyniki: {poprawione_wyniki}")
print(f"Nowa średnia: {nowa_srednia}")

# --- 3.6. Klasyfikacja wyników (np.where) ---
# np.where(warunek, wartość_jeśli_prawda, wartość_jeśli_fałsz)
kategorie = np.where(poprawione_wyniki >= 85, 'bardzo dobry',
            np.where(poprawione_wyniki >= 70, 'dobry',
            np.where(poprawione_wyniki >= 50, 'dostateczny', 'niedostateczny')))

print(f"Kategorie: {kategorie}")

# --- 3.7. Wykrywanie nietypowych wyników ---
srednia_fin = np.mean(poprawione_wyniki)
outliers = poprawione_wyniki[poprawione_wyniki > 1.5 * srednia_fin]

print(f"Średnia końcowa: {srednia_fin}")
print(f"Potencjalne wartości odstające (outliers): {outliers}")
```

```python
import pandas as pd

df = pd.read_excel("Dane_probne.xlsx")


najlepszy_oddzial = df.groupby("oddzial")["miesieczny_koszt"].mean().idxmax()
print(f"Oddział z najwyższym średnim kosztem: {najlepszy_oddzial}")
df_filtered = df[df["oddzial"] == najlepszy_oddzial]
print(f"\nRekordy z oddziału {najlepszy_oddzial}:")
print(df_filtered)
liczba_przedluzonych = df_filtered[df_filtered["czy_przedluzyl"] == True].shape[0]
print(f"\nLiczba osób, które przedłużyły usługę w tym oddziale: {liczba_przedluzonych}")
df_filtered = df_filtered.copy() 
df_filtered["koszt_po_podwyzce_10"] = df_filtered["miesieczny_koszt"] * 1.10
ile_powyzej_1000 = df_filtered[df_filtered["koszt_po_podwyzce_10"] > 1000].shape[0]
print(f"Liczba rekordów z kosztem > 1000 po podwyżce: {ile_powyzej_1000}")
print("\n3 rekordy z najwyższym kosztem po podwyżce:")
print(df_filtered.nlargest(3, "koszt_po_podwyzce_10"))
```
