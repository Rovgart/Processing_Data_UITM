```python
# Zadanie 3.3

# Przyjęty próg zaliczenia
prog = 50

# 1. Wyświetlenie tylko wyników zaliczających (>= 50)
zdane = wyniki_losowe[wyniki_losowe >= prog]

# 2. Wyświetlenie tylko wyników niezaliczających (< 50)
niezdane = wyniki_losowe[wyniki_losowe < prog]

# 3. Liczenie, ilu studentów zdało
liczba_zdanych = zdane.size  # lub len(zdane)

print(f"Wyniki zaliczające: {zdane}")
print(f"Wyniki niezaliczające: {niezdane}")
print(f"Liczba studentów, którzy zdali: {liczba_zdanych}")
```
